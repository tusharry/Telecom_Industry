```python
plt.figure(figsize = (5, 5))
plt.pie (x['gender'].value_counts(), labels = x['gender'].value_counts().keys(), autopct = '%1f%%', startangle = 90)
plt.title('GENDER')
plt.show()

plt.figure(figsize = (5, 5))
plt.pie (x['SeniorCitizen'].value_counts(), labels = ['Retired','Not Retired'], autopct = '%1f%%', startangle = 90)
plt.title('Senior Citizen')
plt.show()
```

```python
plt.figure(figsize = (5, 5))
plt.pie(x['Churn'].value_counts(), labels = ['No', 'Yes'], autopct = '%1f%%', startangle = 90)
plt.title('Churn')
plt.show()
```

```python
plt.figure(figsize = (5, 5))
sns.countplot(x.gender, hue = x.Churn)
plt.show()

plt.figure(figsize = (5, 5))
sns.countplot(x.SeniorCitizen, hue = x.Churn)
plt.show()
```

```python
plt.figure(figsize = (10, 10))
sns.histplot(x['tenure'], bins = 70, kde = True, color = 'red')
plt.grid()
plt.show()
```

```python
for i in x.columns:
    print('Unique values in', i, 'are:', x[i].unique())
    print(''*100)
```

```python
x = x.replace('Female', 0)
x = x.replace('Male', 1)
x = x.replace('No', 0)
x = x.replace('Yes', 1)
x = x.replace('No phone service', 2)
x = x.replace('No internet service', 2)
x = x.replace('Two year', 24)
x = x.replace('Month-to-month', 1)
x = x.replace('One year', 12)
x = x.replace('Fiber optic', 1)
x = x.replace('DSL', 2)
x = x.replace('Credit card (automatic)', 0)
x = x.replace('Bank transfer (automatic)', 1)
x = x.replace('Electronic check', 2)
x = x.replace('Mailed check', 3)
```

```python
for i in x.columns:
    print('Unique values in', i, 'are:', x[i].unique())
    print(''*100)
```
